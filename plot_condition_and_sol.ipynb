# Plot the initial condition and final solution with respect to K.

```python
import numpy as np
import matplotlib.pyplot as plt
from functions import *

N_mus = np.array([1, 5, 10, 20, 40])
N_h = 50
M_h = N_h
h = 1/(N_h + 1)
legend = []
Y_Newmarks = []
Y_Exacts = []
fig = plt.figure(figsize=(5, 4))
for N_mu in N_mus:
    Mus = np.random.uniform(low = -1, high = 1, size = N_mu)
    w = fct(Mus, K = 1)
    cond_init = np.vectorize(w)(np.arange(1, N_h+1)*h)
    plot_sol(N_h, cond_init)
    legend.append(r"$N_\mu = $" + str(N_mu))
    y_Newmark, time =newmark(N_h, M_h, w, h, h)
    Y_Newmarks.append(y_Newmark)
    u_exact = sol_given_fct2(w)
    y_exact = np.vectorize(u_exact)(np.arange(1, N_h+1)*h, 1)
    Y_Exacts.append(y_exact)

plt.legend(legend, bbox_to_anchor=(0.9,1.25), ncol = 3)
#plt.savefig('Figures/Intro/init_cond')
plt.show()

fig = plt.figure(figsize=(5, 4))
legend = []
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
for i, N_mu in enumerate(N_mus):
    plt.plot(np.arange(1, N_h+1)*h, Y_Newmarks[i], 'x' , color = colors[i])
    legend.append(r"$N_\mu = $" + str(N_mu) +", Newmark sol.")
    plt.plot(np.arange(1, N_h+1)*h, Y_Exacts[i], color = colors[i])
    legend.append(r"$N_\mu = $" + str(N_mu) +", Exact sol.")

plt.legend(legend,  bbox_to_anchor=(1.8,1.4), ncol = 3)
#plt.savefig('Figures/Intro/final_sol', bbox_inches='tight')
plt.show()
```
